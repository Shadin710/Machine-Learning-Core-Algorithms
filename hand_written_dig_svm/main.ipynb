```python
import pandas as pd
import numpy as np
import joblib
from sklearn.model_selection import train_test_split
from sklearn import svm
import seaborn as sn
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.metrics import confusion_matrix
```

```python
digit = load_digits()
dir(digit)
```

```python
digit.target_names
```

```python
plt.gray()
plt.matshow(digit.images[0])
```

```python
x_train,x_test,y_train,y_test = train_test_split(digit.data,digit.target,test_size = 0.2)
```

```python
len(x_train)
```

```python
len(y_train)
```

```python
digit_model= svm.SVC(kernel='rbf')
```

```python
digit_model.fit(x_train,y_train)
```

```python
digit_model.score(x_test,y_test)
```

```python
pred = digit_model.predict([digit.data[0]])
```

```python
pred
```

```python
y_pred = digit_model.predict(x_test)
y_pred
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
plt.figure(figsize=(8,4))
sn.heatmap(cm,annot=True)
plt.xlabel("Predicted")
plt.ylabel("Real Values")
```

```python
joblib.dump(digit_model,"digit_model_svm")
```
