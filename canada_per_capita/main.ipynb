```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv")
df
```

```python
y=df['per capita income (US$)']
x = df['year']
y
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],y)
```

```python
reg.predict([[2014]])
```

```python
x.to_csv("eva.csv",columns=['year'],index=False)

df_eva = pd.read_csv("eva.csv")
```

```python
new_df = reg.predict(df_eva)
new_df
```

```python
df_eva['per capita income (US$)'] = new_df
df_eva
```

```python
plt.plot(df['year'],df['per capita income (US$)'],'ro',marker='+')
plt.plot(df_eva['year'],df_eva['per capita income (US$)'])
```
