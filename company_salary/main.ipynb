```python
import pandas as pd
import numpy as np
import seaborn as sn
import joblib 
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("data.csv")
df
```

```python
get_x = df.drop('salary_more_then_100k',axis="columns")
get_x
```

```python
y = df['salary_more_then_100k']
y
```

```python
dummy = pd.get_dummies(df[['company','job','degree']],drop_first=True)
dummy
```

```python
final = pd.concat([get_x,dummy],axis="columns")
final
```

```python
final_x = final.drop(['company','job','degree'],axis="columns")
final_x
```

```python
x_train,x_test,y_train,y_test = train_test_split(final_x,y,test_size =0.3)
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
joblib.dump(model,"company_salary")
```
