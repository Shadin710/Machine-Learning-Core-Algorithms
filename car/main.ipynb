```python
import pandas as pd 
import numpy as np
from sklearn import linear_model
import joblib
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('./data/CarPrice_Assignment.csv')
df
```

## Cleaning the data set

```python
columns =[]
for i in df:
    columns.append(i)
columns
```

```python
df_x = df.drop('price',axis="columns")
df_x
```

```python
dummy  = pd.get_dummies(df_x[['CarName','enginelocation','fueltype','aspiration','carbody','drivewheel','fuelsystem','enginetype']],drop_first = True)
dummy
```

```python
df_x = df_x.drop(['CarName','enginelocation','fueltype','aspiration','carbody','drivewheel','fuelsystem','enginetype'],axis="columns")
df_x
```

```python
final_categorical = pd.concat([df_x,dummy],axis="columns")
final_categorical
```

```python
def is_nan(x):
    return x!=x
```

```python
number  = ['zero','one','two','three','four','five','six','seven','eight','nine','ten','eleven','twelve']

def w2n_doornumber():
    count=0
    get_num=0
    for i in final_categorical.doornumber:
        j=0
        if is_nan(i):
            count= count+1
            continue
        for j in range(len(number)):
            if i==number[j]:
                get_num = j
                break
        final_categorical.doornumber[count]=get_num
        count= count+1
def w2n_cylindernumber():
    count=0
    get_num=0
    for i in final_categorical.cylindernumber:
        j=0
        if is_nan(i):
            count= count+1
            continue
        for j in range(len(number)):
            if i==number[j]:
                get_num = j
                break
        final_categorical.cylindernumber[count]=get_num
        count= count+1 
w2n_cylindernumber()
w2n_doornumber()
```

```python
final_categorical
```

```python
def clean():
    count=0
    
    for i in final_categorical.symboling:  
        if i<0:
            final_categorical.symboling[count]=0
        count=count+1
clean()
final_categorical
```

## The dataset has been cleaned

# Now to train the model

## first we need to split the dataset

```python
y=df['price']
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(final_categorical,y,test_size=0.2)
```

```python
car_model =  linear_model.LinearRegression()
```

```python
car_model.fit(x_train,y_train)
```

```python
car_model.score(x_test,y_test)
```
