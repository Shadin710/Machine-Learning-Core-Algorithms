```python
df = pd.read_csv("./data/IRIS.csv")
df
```

```python
target = df.species.unique()
target[0]
```

```python
df['target']=0
df
```

```python
count=0
for i in df.species:
    if i == 'Iris-setosa':
        df['target'][count]=0
    elif i=='Iris-versicolor':
        df['target'][count]=1
    else:
        df['target'][count]=2
    count = count+1
df
```

```python
get_x=df.drop(['species','target'],axis="columns")
get_x
```

```python
get_y=df['target']
```

```python
x_train,x_test,y_train,y_test= train_test_split(get_x,get_y,test_size=0.2)
```

```python
model = SVC()
```

```python
get_s = cross_val_score(model,get_x,get_y,cv=5)
get_s
```

```python
avg = sum(get_s)/len(get_s)
avg
```

```python
model.fit(x_train,y_train)
```

```python
p = model.predict([[5.1,3.5,1.4,0.3]])
target[p]
```

```python
pred  = model.predict(x_test)
pred
```

```python
flower=target[pred]
flower
```

```python
get_data_frame = pd.DataFrame(flower,columns=['species'])
get_data_frame
```
