```python
df_new =df[['satisfaction_level','average_montly_hours','Work_accident','promotion_last_5years','salary']]
df_new
```

```python
dummy = pd.get_dummies(df_new['salary'],drop_first=True)
dummy
```

```python
final= df_new.drop('salary',axis='columns')
final
```

```python
final =pd.concat([final,dummy],axis="columns")
final
```

```python
x=final
x
```

```python
y= df.left
y.loc[6280]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
x_train
```

```python
model = linear_model.LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
joblib.dump(model,"employee_model")
```
